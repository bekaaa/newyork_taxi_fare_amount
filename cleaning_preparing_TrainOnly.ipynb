```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from datetime import datetime
```

```python
#train = pd.read_csv('./train.csv.zip',compression='zip', 
 #                   dtype={'date': str, 'fullVisitorId': str, 'sessionId':str}, nrows=None)
train = pd.read_hdf('./new_train.hdf', key='date')
```

```python
print(train.info())
train.head(5)
```

```python
totals = train.totals.copy()
cols = ['bounces', 'hits', 'newVisits', 'pageviews', 'visits','transactionRevenue']
totals_df = pd.DataFrame(columns=cols, data=np.zeros((totals.shape[0],len(cols))))

for t,c in zip(totals,range(totals.shape[0])):
    for i in t[1:-1].split(', ') :
        typ,val = [ j[1:-1] for j in i.split(': ')]
        totals_df.at[c,typ] = val

        
new_train = pd.concat((train,totals_df),sort=False,axis=1)
new_train.drop(columns=['totals'], inplace=True)

new_train.to_hdf('./new_train.hdf',key='totals')
```

----------------

## TrafficSource

```python
traffic = train.trafficSource.copy()
```

```python
cols = ['campaign', 'medium', 'adNetworkType', 'gclId', 'adwordsClickInfo', 'targetingCriteria',
        'isTrueDirect', 'adContent', 'isVideoAd', 'slot', 'campaignCode', 'criteriaParameters',
        'referralPath', 'source', 'keyword']

traffic_df = pd.DataFrame(columns=cols, data=np.zeros((traffic.shape[0],len(cols))),dtype=str)


for t,c in zip(traffic,range(traffic.shape[0])):
    for i in t[1:-1].split(', '):
        try :
            typ,val = [ j[1:-1] for j in i.split(': ',1)]
            traffic_df.at[c,typ] = val
        except:
            print(i,end=', ')
print('\n\ns',list(cols))
```

```python
def is_it_worth(S):
    val_counts = S.value_counts()
    index = val_counts.index.tolist()
    values = val_counts.tolist()
    for i in range(len(index)):
        print(index[i], ': ', values[i], '\t', round(values[i]*100./S.shape[0],1),'%')
```

```python
#c=0
c+=1
print(cols[c])
is_it_worth(traffic_df[cols[c]])
```

```python
traffic_df = traffic_df[['medium','isTrueDirect','isVideoAd','slot','source']].copy()

Lencoder = LabelEncoder()
traffic_df.medium = Lencoder.fit_transform(traffic_df.medium)
traffic_df.isTrueDirect = Lencoder.fit_transform(traffic_df.isTrueDirect)
traffic_df.isVideoAd = Lencoder.fit_transform(traffic_df.isVideoAd)
traffic_df.slot = Lencoder.fit_transform(traffic_df.slot)

# later I'll work on this
traffic_df.drop(columns=['source'],inplace=True)

traffic_df.head(3)
```

```python
new_train = pd.concat((new_train,traffic_df),sort=False,axis=1)
new_train.drop(columns=['trafficSource'], inplace=True)

new_train.to_hdf('./new_train.hdf',key='trafficSource')
```

## ChannelGrouping, SocialEngagementType, Date

```python
# ChannelGrouping
Lencoder = LabelEncoder()
new_train.channelGrouping = Lencoder.fit_transform(new_train.channelGrouping)
```

```python
# SocialEngagementType
new_train.drop(columns=['socialEngagementType'], inplace=True)
```

```python
# Date
date = new_train.date.copy()
date_df = pd.DataFrame(columns=['year', 'month', 'dayOfYear'], data=np.zeros((date.shape[0],3)))

for d,ind in zip(date,range(len(date))) :
    date_df.at[ind,'year'] = np.int(d[:4])
    date_df.at[ind,'month'] = np.int(d[4:6])
    adate = datetime.strptime(d, "%Y%m%d")
    date_df.at[ind,'dayOfYear'] = adate.timetuple().tm_yday
```

```python
new_train = pd.concat((new_train,date_df),sort=False,axis=1)
new_train.drop(columns=['date'], inplace=True)

new_train.to_hdf('./new_train.hdf',key='date')
```

## geoNetwork

```python
geoNetwork = new_train.geoNetwork.copy()

cols = ['continent', 'subContinent', 'country']
geoNetwork_df = pd.DataFrame(data=np.zeros([len(geoNetwork),len(cols)]),columns=cols, dtype=np.str)

typs=set()
for t,c in zip(geoNetwork,range(geoNetwork.shape[0])):
    for i in t[1:-1].split(', ') :
        typ,val = [ j[1:-1] for j in i.split(': ')]
        if typ in cols :
            geoNetwork_df.at[c,typ] = val
```

```python
lencoder = LabelEncoder()
geoNetwork_df.continent = lencoder.fit_transform(geoNetwork_df.continent)
geoNetwork_df.subContinent = lencoder.fit_transform(geoNetwork_df.subContinent)
geoNetwork_df.country = lencoder.fit_transform(geoNetwork_df.country)

```

```python
new_train = pd.concat((new_train,geoNetwork_df),sort=False,axis=1)
new_train.drop(columns=['geoNetwork'], inplace=True)

new_train.to_hdf('./new_train.hdf',key='latest')
```

## Device



```python
device = new_train.device.copy()
```

```python

cols = ['browser', 'operatingSystem', 'deviceCategory']
device_df = pd.DataFrame(data=np.zeros([len(device),len(cols)]),columns=cols, dtype=np.str)

typs=set()
for t,ind in zip(device,range(device.shape[0])):
    for i in t[1:-1].split(', ') :
        try:
            typ,val = [ j[1:-1] for j in i.split(': ')]
            if typ in cols :
                device_df.at[ind,typ] = val
        except:
            print(i)
```

```python
freq = device_df.browser.value_counts()
browsers = freq[ freq > 2000].index.tolist()
device_df.loc[~device_df.browser.isin(browsers), 'browser'] = 0

# fix some str/int problem
device_df = device_df.astype(dtype=np.str)

lencoder = LabelEncoder()
device_df.browser = lencoder.fit_transform(device_df.browser)
device_df.operatingSystem = lencoder.fit_transform(device_df.operatingSystem)
device_df.deviceCategory = lencoder.fit_transform(device_df.deviceCategory)
```

```python
new_train = pd.concat((new_train,device_df),sort=False,axis=1)
new_train.drop(columns=['device'], inplace=True)

new_train.to_hdf('./new_train.hdf',key='latest')
```
